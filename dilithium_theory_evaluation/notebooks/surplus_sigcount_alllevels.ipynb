```python
import pandas as pd
```

```python
pd.options.display.width = 0
pd.set_option('display.max_columns', None)
pd.set_option('display.min_rows', 20)
```

```python
ROUND_TO = 55
LEVEL_TO_MAX_M = {2: 1024, 3: 1280, 5: 1792}
```

```python
df_l2 = pd.read_json('l2_dilithium_surplus_sigcount_total.json').add_prefix('l2_').round({'l2_percent_x': ROUND_TO})
df_l2['l2_f'] = LEVEL_TO_MAX_M[2] - df_l2.index 
df_l2
```

```python
l2_max_percent_nosurplus = df_l2[df_l2['l2_surplus'] == 0]['l2_percent_x'].max()
l2_max_percent_nosurplus
```

```python
df_l3 = pd.read_json('l3_dilithium_surplus_sigcount_total.json').add_prefix('l3_').round({'l3_percent_x': ROUND_TO})
df_l3['l3_f'] = LEVEL_TO_MAX_M[3] - df_l3.index 
df_l3
```

```python
l3_max_percent_nosurplus = df_l3[df_l3['l3_surplus'] == 0]['l3_percent_x'].max()
l3_max_percent_nosurplus
```

```python
df_l5 = pd.read_json('l5_surplus_sigcount_total.json').add_prefix('l5_').round({'l5_percent': ROUND_TO})
df_l5['l5_f'] = LEVEL_TO_MAX_M[5] - df_l5.index 
df_l5['l5_percent_x'] = df_l5['l5_percent']
l5_max_percent_nosurplus = df_l5[df_l5['l5_surplus'] == 0]['l5_percent'].max()
l5_max_percent_nosurplus
df_l5
```

```python
min_percent = min(l3_max_percent_nosurplus, l2_max_percent_nosurplus, l5_max_percent_nosurplus)
min_percent
```

```python
min_percent = .99
```

```python
l2_upper = df_l2[df_l2['l2_surplus'] > 0]
l3_upper = df_l3[df_l3['l3_surplus'] > 0]
l5_upper = df_l5[df_l5['l5_surplus'] > 0]
```

```python
l2_upper
```

```python
l3_upper
```

```python
l5_upper
```

```python
joined = l2_upper.set_index('l2_f').join(l3_upper.set_index('l3_f'), how='outer').join(l5_upper.set_index('l5_f'), how='outer')
joined.index.name = 'f'
joined
```

```python
ax = joined[['l5_num_signatures', 'l3_num_signatures', 'l2_num_signatures']].rename(columns={
        'l2_num_signatures': 'NIST Security Level II',
        'l3_num_signatures': 'NIST Security Level III',
        'l5_num_signatures': 'NIST Security Level IV'}
                                                         ).plot(kind='bar')
ax.set_xlabel('number of zero-coefficients (f)')
ax.set_ylabel('average signature count')
fig = ax.get_figure()
fig.savefig(f'dilithium_sigcount_upper.pdf', format="pdf", bbox_inches='tight')
ax
# if for a security level a bar is not present it means that for for that security level
# the surplus of equations needed was zero. 
```

```python
l2_lower = df_l2[df_l2['l2_surplus'] == 0]
l3_lower = df_l3[df_l3['l3_surplus'] == 0]
l5_lower = df_l5[df_l5['l5_surplus'] == 0]
```

```python
ljoined = l2_lower.set_index('l2_percent_x').join(l3_lower.set_index('l3_percent_x'), how='outer').join(l5_lower.set_index('l5_percent'), how='outer').fillna(method='bfill')
#joined.index.name = 'percent'
ljoined
```

```python
max_percent_l2 = ljoined['l2_percent_y'].max()
max_percent_l3 = ljoined['l3_percent_y'].max()
max_percent_l5 = ljoined['l5_percent_x'].max()
max_percent_l2, max_percent_l3, max_percent_l5
```

```python
ax = ljoined[['l5_num_signatures', 'l3_num_signatures', 'l2_num_signatures']].rename(columns={
        'l2_num_signatures': 'NIST Security Level II',
        'l3_num_signatures': 'NIST Security Level III',
        'l5_num_signatures': 'NIST Security Level IV'}
                                                         ).plot()
ax.set_xlabel('no. of iterations before fault occurs (%)')
ax.set_ylabel('average signature count')
ax.plot(max_percent_l2, ljoined.loc[max_percent_l2]['l2_num_signatures'], 'x', color='tab:green')
ax.plot(max_percent_l3, ljoined.loc[max_percent_l3]['l3_num_signatures'], 'x', color='tab:orange')
ax.plot(max_percent_l5, ljoined.loc[max_percent_l5]['l5_num_signatures'], 'x', color="tab:blue")
fig = ax.get_figure()
fig.savefig(f'dilithium_sigcount_lower.pdf', format="pdf", bbox_inches='tight')
ax
```
